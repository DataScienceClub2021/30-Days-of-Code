```python
class BST:
  def __init__(self,val=None):
    self.left = None
    self.right =None 
    self.val  = val
  def insert(self,val):
    if not self.val :
      self.val = val
      return
    if self.val == val:
      return 
    if val < self.val:
      if self.left:
        self.left.insert(val)
        return 
      self.left = BST(val)
      return 
    else :
      if self.right:
        self.right.insert(val)
        return 
      self.right = BST(val)
      return 
  def preorder(self,ls): # Rlr
    if self.val is not None : 
      ls.append(self.val)
    if self.left is not None:
      self.left.preorder(ls)
    if self.right is not None:
      self.right.preorder(ls)
    return ls
  def postorder(self,ls): #lrR
    if self.left is not None:
      self.left.postorder(ls)
    if self.right is not None:
      self.right.postorder(ls)
    if self.val is not None:
      ls.append(self.val)
    return ls
  def inorder(self,ls): #lRr
    if self.left is not None:
      self.left.inorder(ls)
    if self.val is not None:
      ls.append(self.val)
    if self.right is not None:
      self.right.inorder(ls)
    return ls

    
```

```python
BST_node = BST()
BST_node.insert(10)
BST_node.insert(5)
BST_node.insert(10)
BST_node.insert(20)
BST_node.insert(30)
BST_node.insert(40)
BST_node.insert(50)
```
